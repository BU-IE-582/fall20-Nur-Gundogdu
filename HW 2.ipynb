```python
import scipy.spatial.distance
from sklearn.metrics import euclidean_distances
from sklearn.datasets import load_digits
from sklearn.manifold import MDS
from sklearn import manifold
```

```python
frames= [X_train_df2,Y_train_df2,Z_train_df2]
newdf = pd.concat(frames, axis =1)

distX = euclidean_distances(X_train_df2)
distY = euclidean_distances(Y_train_df2)
distZ = euclidean_distances(Z_train_df2)

finaldist = distX + distY + distZ

#symmetric distance matrix
M2 = pd.DataFrame(finaldist)
M2
```

```python
mds = manifold.MDS(2, dissimilarity='precomputed')
XY = mds.fit_transform(M2)
XY
```
